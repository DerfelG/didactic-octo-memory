# Indice de confianca do empresário industrial com as condições atuais

```python
a=ipeadatapy.timeseries("CNI12_ICEIGER12").reset_index()
date_time = pd.to_datetime(a.DATE)
a = a.set_index(date_time)
a
```

```python
sns.set(font_scale=1.4)
sns.set(rc={'figure.figsize':(11, 4)})
ax = a.loc['2018':'2021', "VALUE (-)"].plot(color='green', linewidth=1.5, marker='o', markersize=5, linestyle='-')
ax.set_ylabel('Índice')
plt.xlabel("Data", labelpad=10)
plt.ylabel("ICEI", labelpad=10)
plt.title("Índice de confiança do empresário industrial (ICEI): condições atuais", y=1.02, fontsize=10);
plt.savefig("Empresário Industrial")
```

# Divida bruta

```python
db=pd.read_csv("STI-20210612223301113.csv", sep=";")
db.columns.values[1] = "DBGG"
db=db[0:174]
date_time = pd.to_datetime(db.Date)
db=db.set_index(date_time)
db["DBGG"] = pd.to_numeric(db["DBGG"])
db

```

```python
i=["2014-12", "2015-12", "2016-12", "2017-12", "2018-12", "2019-12", "2020-12", "2021-12"]
anos=["2014","2015","2016", "2017", "2018", "2019", "2020", "2021*"]
sns.set(rc={'figure.figsize':(11, 4)})
plt.title("Divida bruta do governo federal - % PIB", y=1.02, fontsize=15);

for h in i:
    ax = db.loc[i, 'DBGG'].plot.bar(color='olive')
    ax.set_xticklabels(anos, rotation=0)
    ax.set_ylabel('Divida Bruta - % PIB')
    ax.set_xlabel("")

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.015, p.get_height() * 1.005))
plt.savefig("Divida")


    
```
